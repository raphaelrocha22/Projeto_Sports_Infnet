Documentação API: https://www.api-football.com/documentation-v3

```python
#Exibir totas as colunas no pandas
pd.set_option('display.max_columns', None)

season = 2023
base_url = "https://v3.football.api-sports.io"
headers = {
  'x-rapidapi-key': '75a1d666035ab70c3321c2884333be2b',
  'x-rapidapi-host': 'v3.football.api-sports.io'
}
```

Chamada API Brasileirão 2023

```python
url_leagues = f"{base_url}/leagues?country=Brazil&season=2023&name=Serie A"
response_leagues = requests.get(url_leagues, headers=headers).json()
```

```python
response_leagues
```

```python
#identificação do ID do Brasileirão serie A
id_liga = response_leagues['response'][0]['league']['id']
print(id_liga)
```

Consulta das partidas do campeonato para capturar id

```python
url_partidas = f"{base_url}/fixtures?timezone=America/Sao_Paulo&league={id_liga}&season={season}"
response_partidas = requests.get(url_partidas,headers=headers).json()
```

```python
response_partidas
```

```python
df_partidas = pd.json_normalize(response_partidas['response'])
df_partidas['fixture.date'] = pd.to_datetime(df_partidas['fixture.date'])
df_partidas
```

Analise de Dados: filtrando algumas partidas

Partida de Botafogo x Vasco - 06/11/23

```python
#Filtrando partidas de 06/11/23
df_partidas[df_partidas['fixture.date'].dt.date.astype(str) == "2023-11-06"]
```

```python
#Resultado Jogo Bota x Vasco
df_partidas[df_partidas['fixture.id'] == 1005960]\
    [['teams.home.name', 'goals.home', 'teams.away.name', 'goals.away']]
```

```python
#Consulta estatisticas da partida
url_statistics = f"{base_url}/fixtures/statistics?fixture=1005960"
response_statistics = requests.get(url_statistics, headers=headers).json()
```

```python
response_statistics
```

```python
df_estatistica = pd.json_normalize(response_statistics['response'],
                                   record_path="statistics",
                                   meta=[["team", "id"], ["team", "name"]]
                                )
```

```python
#Estatisticas do Jogo
df_estatistica.pivot(index="type", columns="team.name", values="value")
```
